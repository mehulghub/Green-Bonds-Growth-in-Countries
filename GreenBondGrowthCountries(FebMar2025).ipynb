```python
from IPython import get_ipython
from IPython.display import display
import pandas as pd
from google.colab import files
import matplotlib.pyplot as plt
from datetime import datetime
import yfinance as yf
import numpy as np
import statsmodels.api as sm
from scipy import stats
import seaborn as sns

class GreenBondsAnalyzer:
    """
    A class to analyze Green Bonds data.
    """
    def __init__(self, excel_file, csv_file):
        """
        Initializes the GreenBondsAnalyzer with file paths.

        Args:
            excel_file (str): Path to the Excel file.
            csv_file (str): Path to the CSV file.
        """
        self.excel_file = excel_file
        self.csv_file = csv_file
        self.green_bonds_df = None

    def load_and_preprocess_data(self):
        """
        Loads data from Excel and CSV, handles date format inconsistency,
        fills missing values, and cleans column names.
        """
        # Load data from Excel
        self.green_bonds_df = pd.read_excel(self.excel_file, dtype={'Amount Issued': str})
        print("Initial Green Bonds DataFrame from Excel:")
        display(self.green_bonds_df)
        self.green_bonds_df.info()

        # Load data from CSV to fix date formats
        green_bonds_df_csv = pd.read_csv(self.csv_file)
        print("\nGreen Bonds DataFrame from CSV (for date correction):")
        display(green_bonds_df_csv.head())

        # Fix date formats if lengths match
        if len(self.green_bonds_df) == len(green_bonds_df_csv):
            self.green_bonds_df['Issue Date'] = green_bonds_df_csv['Issue Date']
            self.green_bonds_df['Maturity Date'] = green_bonds_df_csv['Maturity Date']
            print("\nGreen Bonds DataFrame with corrected dates:")
            display(self.green_bonds_df.head())
        else:
            print("\nWarning: Excel and CSV file lengths do not match. Date correction skipped.")

        # Fill missing values
        self.green_bonds_df.fillna(0, inplace=True)
        print("\nDataFrame after filling NaNs:")
        display(self.green_bonds_df.head())
        print("\nMissing values per column after filling:")
        print(self.green_bonds_df.isnull().sum())

        # Clean column names
        self.green_bonds_df.columns = (
            self.green_bonds_df.columns.str.lower()
            .str.replace(" ","_")
            .str.replace("[^a-z0-9]","", regex = True)
        )
        print("\nCleaned column names:")
        print(self.green_bonds_df.columns.tolist())

    def check_duplicates(self):
        """
        Checks for duplicate rows in the DataFrame.
        """
        duplicates = self.green_bonds_df.duplicated()
        print("\nNumber of duplicate rows:", duplicates.sum())

    def map_currency_to_country(self):
        """
        Maps currency codes to country names and creates a new 'country' column.
        """
        country_map = {
            'JPY': 'japan', 'USD': 'unitedstates', 'MYR': 'malaysia',
            'KRW': 'southkorea', 'CNY': 'china', 'EUR': 'eurozone',
            'TWD': 'taiwan', 'AUD': 'australia', 'THB': 'thailand',
            'SEK': 'sweden', 'IDR': 'indonesia', 'CAD': 'canada',
            'SGD': 'singapore', 'NOK': 'norway', 'GBP': 'unitedkingdom',
            'HKD': 'hongkong', 'DKK': 'denmark', 'CHF': 'switzerland',
            'PLN': 'poland', 'ISK': 'iceland', 'INR': 'india',
            'KZT': 'kazakhstan'
        }
        self.green_bonds_df['country'] = self.green_bonds_df['currency'].map(country_map).fillna('Unknown')
        print("\nDataFrame after mapping currency to country:")
        self.green_bonds_df.info()
        display(self.green_bonds_df.head())

    def correct_maturity_date_format(self):
        """
        Corrects the format of the 'maturitydate' column.
        """
        self.green_bonds_df['maturitydate_split'] = self.green_bonds_df['maturitydate'].astype(str).str.split('-')
        months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
        for index, row in self.green_bonds_df.iterrows():
            if isinstance(row['maturitydate_split'], list) and row['maturitydate_split'][0] in months:
                self.green_bonds_df.loc[index, 'maturitydate'] = row['maturitydate_split'][1] + '-' + row['maturitydate_split'][0]
        self.green_bonds_df = self.green_bonds_df.drop(columns=['maturitydate_split'])
        print("\nDataFrame after correcting maturity date format:")
        display(self.green_bonds_df.head())

    def convert_amount_to_usd(self, exchange_rates):
        """
        Converts 'amountissued' to USD based on provided exchange rates.

        Args:
            exchange_rates (dict): A dictionary of exchange rates to USD.
        """
        self.green_bonds_df['amountissued'] = pd.to_numeric(self.green_bonds_df['amountissued'], errors='coerce')
        self.green_bonds_df['amountissued_usd'] = self.green_bonds_df.apply(
            lambda row: row['amountissued'] * exchange_rates.get(row['currency'], 0) if pd.notna(row['amountissued']) else 0,
            axis=1
        )
        print("\nDataFrame with amount issued in USD:")
        print(self.green_bonds_df[['currency', 'amountissued', 'amountissued_usd']])
        self.green_bonds_df.info()

    def analyze_entity_issuance(self, top_n=10):
        """
        Analyzes green bonds issuance by entity and plots the top N entities.

        Args:
            top_n (int): The number of top entities to display and plot.
        """
        entity_amountissued_sum = self.green_bonds_df.groupby('entity')['amountissued_usd'].sum().reset_index()
        entity_amountissued_sum_sorted = entity_amountissued_sum.sort_values(by='amountissued_usd', ascending=False)
        print(f"\nSum of Amount Issued (USD) by Entity (Sorted):")
        display(entity_amountissued_sum_sorted.head())

        top_entity_green_bonds = entity_amountissued_sum_sorted.head(top_n)
        print(f"\nTop {top_n} Entities for Green Bonds Issuance (Feb-Mar 2025):")
        display(top_entity_green_bonds)

        plt.figure(figsize=(12, 6))
        plt.bar(top_entity_green_bonds['entity'], top_entity_green_bonds['amountissued_usd'])
        plt.title(f'Top {top_n} Entities for Green Bonds Issuance (Feb-Mar 2025)')
        plt.xlabel('Entity')
        plt.ylabel('Amount Issued (USD)')
        plt.xticks(rotation=45, ha='right')
        plt.tight_layout()
        plt.show()

        top_entities = entity_amountissued_sum_sorted.head(10)['entity'].tolist()
        top_entities_df = self.green_bonds_df[self.green_bonds_df['entity'].isin(top_entities)]
        print("\nCountry for the top 10 entities by Amount Issued (USD):")
        print(top_entities_df[['entity', 'country']].drop_duplicates().to_string(index=False))

    def analyze_country_issuance(self, top_n=10):
        """
        Analyzes green bonds issuance by country and plots the top N countries.

        Args:
            top_n (int): The number of top countries to display and plot.
        """
        country_amountissued_sum = self.green_bonds_df.groupby('country')['amountissued_usd'].sum().reset_index()
        country_amountissued_sum_sorted = country_amountissued_sum.sort_values(by='amountissued_usd', ascending=False)
        print(f"\nSum of Amount Issued (USD) by Country (Sorted):")
        display(country_amountissued_sum_sorted)

        top_country_green_bonds = country_amountissued_sum_sorted.head(top_n)
        print(f"\nTop {top_n} Countries for Green Bonds Issuance (Feb-Mar 2025):")
        display(top_country_green_bonds)

        plt.figure(figsize=(10, 6))
        plt.bar(top_country_green_bonds['country'], top_country_green_bonds['amountissued_usd'])
        plt.title(f'Top {top_n} Countries for Green Bonds Issuance (Feb-Mar 2025)')
        plt.xlabel('Country')
        plt.ylabel('Amount Issued (USD)')
        plt.xticks(rotation=45, ha='right')
        plt.tight_layout()
        plt.show()

    def analyze_eurozone_entities(self):
        """
        Analyzes green bonds issuance by entities within the Eurozone.
        """
        eurozone_entities = self.green_bonds_df[self.green_bonds_df['country'] == 'eurozone']
        eurozone_entity_amountissued = eurozone_entities.groupby('entity')['amountissued_usd'].sum().reset_index()
        eurozone_entity_amountissued_sorted = eurozone_entity_amountissued.sort_values(by='amountissued_usd', ascending=False)
        print("\nGreen Bonds Issuance by Entity in Eurozone:")
        display(eurozone_entity_amountissued_sorted)

    def conduct_event_study(self, firm_ticker, market_ticker, event_date, estimation_window_start, estimation_window_end, event_window_start, event_window_end):
        """
        Conducts an event study for a given firm and market index.

        Args:
            firm_ticker (str): The ticker symbol for the firm.
            market_ticker (str): The ticker symbol for the market index.
            event_date (str): The date of the event (YYYY-MM-DD).
            estimation_window_start (str): Start date of the estimation window (YYYY-MM-DD).
            estimation_window_end (str): End date of the estimation window (YYYY-MM-DD).
            event_window_start (str): Start date of the event window (YYYY-MM-DD).
            event_window_end (str): End date of the event window (YYYY-MM-DD).
        """
        # Download firm data
        data_firm = yf.download(firm_ticker, start=estimation_window_start, end=event_window_end, auto_adjust=False)
        data_firm['Firm_DailyReturn'] = data_firm['Adj Close'].pct_change() * 100
        print(f"\nFirm Data ({firm_ticker}):")
        display(data_firm)

        # Download market index data
        data_market = yf.download(market_ticker, start=estimation_window_start, end=event_window_end, auto_adjust=False)
        data_market['MarketIndex_DailyReturn'] = data_market['Adj Close'].pct_change() * 100
        print(f"\nMarket Index Data ({market_ticker}):")
        display(data_market)

        # Filter data for estimation window
        data_firm_est = data_firm[(data_firm.index >= estimation_window_start) & (data_firm.index <= estimation_window_end)].copy()
        data_market_est = data_market[(data_market.index >= estimation_window_start) & (data_market.index <= estimation_window_end)].copy()

        # Merge data for estimation window
        merged_data_est = pd.merge(
            data_firm_est[['Firm_DailyReturn']],
            data_market_est[['MarketIndex_DailyReturn']],
            left_index=True,
            right_index=True,
            how='inner'
        ).dropna()
        print("\nMerged Data for Estimation Window:")
        display(merged_data_est)

        # Run Market Model regression
        X = sm.add_constant(merged_data_est['MarketIndex_DailyReturn'])
        y = merged_data_est['Firm_DailyReturn']
        model = sm.OLS(y, X).fit()
        print("\nMarket Model Regression Results:")
        print(model.summary())

        # Extract alpha and beta
        alpha_firm = model.params['const']
        beta_firm = model.params['MarketIndex_DailyReturn']
        print(f"\nEstimated Alpha: {alpha_firm:.6f}")
        print(f"Estimated Beta: {beta_firm:.6f}")

        # Filter data for event window
        data_firm_event = data_firm[(data_firm.index >= event_window_start) & (data_firm.index <= event_window_end)].copy()
        data_market_event = data_market[(data_market.index >= event_window_start) & (data_market.index <= event_window_end)].copy()

        # Merge data for event window
        merged_data_event = pd.merge(
            data_firm_event[['Firm_DailyReturn']],
            data_market_event[['MarketIndex_DailyReturn']],
            left_index=True,
            right_index=True,
            how='inner'
        ).dropna()
        print("\nMerged Data for Event Window:")
        display(merged_data_event)

        # Calculate Expected Returns and Abnormal Returns
        merged_data_event['Expected_Return'] = alpha_firm + beta_firm * merged_data_event['MarketIndex_DailyReturn']
        merged_data_event['Abnormal_Return'] = merged_data_event['Firm_DailyReturn'] - merged_data_event['Expected_Return']
        print("\nEvent Window Data with Abnormal Returns:")
        display(merged_data_event)

        # Calculate Cumulative Abnormal Return (CAR)
        car = merged_data_event['Abnormal_Return'].sum()
        print(f"\nCumulative Abnormal Return (CAR): {car:.4f}")

        # Significance Testing
        self.perform_significance_tests(merged_data_event['Abnormal_Return'])

    def perform_significance_tests(self, abnormal_returns):
        """
        Performs normality test and significance tests on abnormal returns.

        Args:
            abnormal_returns (pd.Series): Series of abnormal returns.
        """
        print("\n--- Significance Testing ---")

        # Normality Test (Shapiro-Wilk)
        shapiro_test_statistic, shapiro_p_value = stats.shapiro(abnormal_returns)
        print(f"\nShapiro-Wilk Test Statistic: {shapiro_test_statistic:.4f}")
        print(f"Shapiro-Wilk p-value: {shapiro_p_value:.4f}")
        alpha = 0.05
        if shapiro_p_value > alpha:
            print("The 'ar' values appear to be normally distributed (fail to reject H0).")
        else:
            print("The 'ar' values do not appear to be normally distributed (reject H0).")

        # Visualize Distribution
        plt.figure(figsize=(10, 5))
        plt.subplot(1, 2, 1)
        sns.histplot(abnormal_returns, kde=True)
        plt.title('Distribution of Abnormal Returns (AR)')
        plt.xlabel('Abnormal Returns')
        plt.ylabel('Frequency')
        plt.subplot(1, 2, 2)
        stats.probplot(abnormal_returns, dist="norm", plot=plt)
        plt.title('QQ Plot of Abnormal Returns (AR)')
        plt.tight_layout()
        plt.show()

        # Parametric t-test
        t_statistic, p_value_t = stats.ttest_1samp(abnormal_returns, 0)
        print(f"\nParametric t-test:")
        print(f"T-statistic: {t_statistic:.4f}")
        print(f"P-value: {p_value_t:.4f}")
        if p_value_t < alpha:
            print("Reject H0: The average abnormal return is significantly different from 0.")
        else:
            print("Fail to reject H0: There is no significant evidence that the average abnormal return is different from 0.")

        # Non-Parametric Chi-Square Test
        abnormal_returns_sign = np.where(abnormal_returns > 0, 'Positive', 'Non-Positive')
        observed_frequencies = pd.Series(abnormal_returns_sign).value_counts()
        print("\nObserved frequencies of AR sign:")
        print(observed_frequencies)
        total_observations = len(abnormal_returns)
        expected_frequencies = pd.Series([total_observations / 2, total_observations / 2], index=['Positive', 'Non-Positive'])
        expected_frequencies = expected_frequencies.reindex(observed_frequencies.index, fill_value=0)

        if (expected_frequencies < 5).any():
            print("\nWarning: Expected frequencies are less than 5. The Chi-square test results may be unreliable.")
            print("Consider using Fisher's Exact Test for small sample sizes.")

        chi2_statistic, p_value_chi2 = stats.chisquare(f_obs=observed_frequencies, f_exp=expected_frequencies)
        print(f"\nNon-Parametric Chi-square Test:")
        print(f"Chi-square statistic: {chi2_statistic:.4f}")
        print(f"P-value: {p_value_chi2:.4f}")
        if p_value_chi2 < alpha:
            print("Reject H0: The distribution of positive/non-positive abnormal returns is significantly different from a 50/50 split.")
        else:
            print("Fail to reject H0: There is no significant evidence that the distribution of positive/non-positive abnormal returns is different from a 50/50 split.")

# --- Usage ---

# Define file paths
excel_file = "GreenBonds(FebMar2025).csv.xlsx"
csv_file = "GreenBonds(FebMar2025).csv"

# Initialize the analyzer
analyzer = GreenBondsAnalyzer(excel_file, csv_file)

# Load and preprocess data
analyzer.load_and_preprocess_data()

# Check for duplicates
analyzer.check_duplicates()

# Map currency to country
analyzer.map_currency_to_country()

# Correct maturity date format
analyzer.correct_maturity_date_format()

# Define exchange rates (replace with actual rates)
exchange_rates_to_usd = {
    'USD': 1.0, 'JPY': 0.0070, 'MYR': 0.24, 'KRW': 0.00073, 'CNY': 0.14,
    'EUR': 1.14, 'TWD': 0.033, 'AUD': 0.65, 'THB': 0.031, 'SEK': 0.11,
    'IDR': 0.000062, 'CAD': 0.73, 'SGD': 0.78, 'NOK': 0.099, 'GBP': 1.36,
    'HKD': 0.13, 'DKK': 0.15, 'CHF': 1.22, 'PLN': 0.27, 'ISK': 0.0079,
    'INR': 0.012, 'KZT': 0.0020
}

# Convert amount issued to USD
analyzer.convert_amount_to_usd(exchange_rates_to_usd)

# Analyze issuance by entity
analyzer.analyze_entity_issuance(top_n=15)

# Analyze issuance by country
analyzer.analyze_country_issuance(top_n=10)

# Analyze Eurozone entities
analyzer.analyze_eurozone_entities()

# Conduct Event Study (Example for Lloyds)
# Replace with actual dates and ticker if needed
firm_ticker = 'LLOY.L'  # Lloyds Banking Group PLC
market_ticker = '^STOXX' # STOXX 600
event_date = '2025-02-15' # Assumed event date
estimation_window_start = '2024-06-10' # 250 days before event
estimation_window_end = '2024-12-31' # 45 days before event
event_window_start = pd.to_datetime(event_date) - pd.Timedelta(days=6) # 5 days before event
event_window_end = pd.to_datetime(event_date) + pd.Timedelta(days=7) # 6 days after event

analyzer.conduct_event_study(
    firm_ticker,
    market_ticker,
    event_date,
    estimation_window_start,
    estimation_window_end,
    event_window_start,
    event_window_end
)
```
